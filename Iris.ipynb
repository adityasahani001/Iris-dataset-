```python
import kagglehub
path = kagglehub.dataset_download("uciml/iris")
print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
file_path = f"{path}/Iris.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
sns.countplot(x='Species', data=df, palette='pastel')
sns.countplot(x='Species', data=df, hue='Species', palette=['red', 'blue', 'green'], legend=False)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
sns.heatmap(df.select_dtypes(include=['number']).corr(method='pearson').drop(
['Id'], axis=1).drop(['Id'], axis=0),
            annot = True);
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.countplot(x='Species', data=df,palette=['red', 'blue', 'green'])
plt.title('Distribution of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```
